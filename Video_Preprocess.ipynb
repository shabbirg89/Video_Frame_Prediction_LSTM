```python
import os 
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
No_of_frames = 15
```

```python
video_path = "/content/drive/MyDrive/Video_Prediction/video/"
files = [video_path + x for x in os.listdir(video_path) if ".avi" in x]
#files = np.random.choice(files, 100)
files = list(set(files))
print("number of videos:",len(files))
```

```python
importedvideos = []
# load all the images and prepare them for feeding into the CNN
counter = 0

for video in files:
    
      frames = []
      count = 0

      path = video
      cap = cv2.VideoCapture(path)

      length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
      print("The name of the file is :",video)
      print("The total frames in the video is :",length)
      temp = length / No_of_frames
      
      ret = True
      while ret:
        ret, img = cap.read() # read one frame from the 'capture' object; img is (H, W, C)
        if ret:
            img = cv2.resize(img, (64, 64))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            frames.append(img)
    #       cv2_imshow(img)
    #       print("****************************")
    #        print("****************************")
            count += temp
            cap.set(1,count)
        else:
          cap.release()
          break
      video_frames = np.expand_dims(np.stack(frames, axis=0),axis=0) 
      print(video_frames.shape)
      importedvideos.append(video_frames) 
      # extract the images features
result = [i for i in importedvideos if i.shape[1] == 15]
videos_train = np.vstack(result)
```
